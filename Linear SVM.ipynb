```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings('ignore')
```

```python
news = pd.read_csv("https://raw.githubusercontent.com/austinyu/CSC395-Final-Project/main/dataset_processed/fakeNews_titleStemming_Sampled.csv")
news
```

```python
from sklearn.model_selection import train_test_split
from sklearn import datasets
from sklearn import svm

news_predictors = news.drop(columns = ['label'])
news_target = news['label']
```

```python
clf = svm.SVC(kernel = "linear") #SVM Model, default C-value
clf.fit(news_predictors, news_target)
accuracy = clf.score(news_predictors, news_target.values.ravel())
print("This is the accuracy for default C-value", accuracy)

C_values = [0.001, 0.01, 0.1, 1] #test various C-values to determine which gives the best accuraccy
for i in C_values:
  clf = svm.SVC(kernel = "linear", C=i)
  clf.fit(news_predictors, news_target)
  print(i, clf.score(news_predictors, news_target.values.ravel()))
```
