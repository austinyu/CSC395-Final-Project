```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings('ignore')
```

```python
news_raw = pd.read_csv('./dataset_processed/fakeNews_languageDetection.csv', index_col=0)
news_raw.drop(columns=['id'], inplace=True)
news_raw.reset_index(inplace=True)
news_raw_predictors = news_raw.drop(columns = ['label'])
news_raw_target = news_raw['label']

news_raw
```

```python
samples_selected = 1000

news_sampled = news_raw.sample(n= samples_selected, random_state=1)
news_sampled.reset_index(inplace=True)
```

```python
from nltk.stem import PorterStemmer
ps = PorterStemmer()

title_parsed = [sentence.split(' ') 
                    for sentence in news_sampled.title.to_numpy()]
for sentence_index, sentence in enumerate(title_parsed):
    for word_index, word in enumerate(sentence):
        title_parsed[sentence_index][word_index] = ps.stem(word)

news_sampled['title'] = [' '.join(sentence) for sentence in title_parsed]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
corpus = news_sampled.title
vectorizer = CountVectorizer()
counter_per_line_sparse = vectorizer.fit_transform(corpus)
counter_per_line = counter_per_line_sparse.toarray()
words = vectorizer.get_feature_names_out()
```

```python
title_stemmed = pd.DataFrame(columns=words, data = counter_per_line, dtype=int)
title_stemmed['label'] = news_sampled.label
```

```python
title_stemmed.to_csv('dataset_processed/fakeNews_titleStemming_Sampled.csv')
```
